```python
import librosa
import matplotlib.pyplot as plt
from scipy.signal import get_window
from scipy import signal 
from librosa.filters import mel
from numpy.random import RandomState
import soundfile as sf
import numpy as np
from pysptk import sptk
import os
from tqdm import tqdm
import pickle
```

```python
def speaker_normalization(f0, index_nonzero, mean_f0, std_f0):
    # f0 is logf0
    f0 = f0.astype(float).copy()
    #index_nonzero = f0 != 0
    f0[index_nonzero] = (f0[index_nonzero] - mean_f0) / std_f0 / 4.0
    f0[index_nonzero] = np.clip(f0[index_nonzero], -1, 1)
    f0[index_nonzero] = (f0[index_nonzero] + 1) / 2.0
    
    
    return f0
```

```python
def butter_highpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='high', analog=False)
    return b, a
```

```python
def pySTFT(x, length, fft_length=1024):
    # print(x.shape)
    x = np.pad(x, int(fft_length // 2), mode='reflect')
    # print(x.shape)
    # print(x.shape[-1])
    hop_length = (x.shape[-1] - fft_length) // (length-1)

    shape = x.shape[:-1] + (length, fft_length)
    # print(shape)
    strides = x.strides[:-1] + (hop_length * x.strides[-1], x.strides[-1])
    # print(len(strides))
    result = np.lib.stride_tricks.as_strided(x, shape=shape,
                                             strides=strides)
    # print(result.shape)

    fft_window = get_window('hann', fft_length, fftbins=True)
    # print(fft_window.shape)
    result = np.fft.rfft(fft_window * result, n=fft_length).T
    # print(result.shape)

    return np.abs(result)
```

```python
def extract_f0_func_audiofile(base, audio, gender='M'):
    floor_sp, ceil_sp = -80, 30
    mel_basis = mel(sr= 16000, n_fft= 1024, fmin=90, fmax=7600, n_mels=80).T
    # print(mel_basis.shape)
    min_level = np.exp(-100 / 20 * np.log(10))
    b, a = butter_highpass(30, 16000, order=5)

    if gender == 'M':
        lo, hi = 50, 250
    elif gender == 'F':
        lo, hi = 100, 600
    else:
        raise ValueError
    prng = RandomState(0)
    # x, fs = sf.read(audio_file)
    audio_file = base + audio

    x, fs = librosa.load(audio_file, sr= 16000)
    # print(f'x: {x.shape}')
    # print(f'fs: {fs}')

    txt_file = base + audio[:-4] + '.txt'
    # print(txt_file)
    with open(txt_file, 'r') as f:
        strings = f.readlines()
        # print(len(strings))

    if(len(x.shape) >= 2):
        x = x[:, 0]
    if x.shape[0] % 256 == 0:
        x = np.concatenate((x, np.array([1e-06])), axis=0)
    y = signal.filtfilt(b, a, x)
    # print(f'y: {y.shape}')
    wav = y * 0.95 + (prng.rand(y.shape[0]) - 0.5) * 1e-06
    # print(wav.shape)
    D = pySTFT(wav, len(strings)).T
    # print(f'D: {D.shape}')
    D_mel = np.dot(D, mel_basis)
    D_db = 20 * np.log10(np.maximum(min_level, D_mel)) - 16
    S = (D_db + 100) / 100

    f0_rapt = sptk.rapt(wav.astype(np.float32) * 32768, fs, 256, min=lo, max=hi, otype=2)
    index_nonzero = (f0_rapt != -1e10)
    tmp = f0_rapt[index_nonzero]
    mean_f0, std_f0 = np.mean(tmp), np.std(tmp)

    f0_norm = speaker_normalization(f0_rapt, index_nonzero, mean_f0, std_f0)

    return S, f0_norm
```
